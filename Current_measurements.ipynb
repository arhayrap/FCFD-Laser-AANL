```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
bias_voltage = np.array(list(np.linspace(1,10,10))+[15,20,25,30]+list(np.linspace(35,60,6)))
# voltage_drop = np.array([2.65,2.7,2.7,2.8,2.9,2.9,3.0,3.1,3.3,3.5,3.7,3.7,3.7,3.3,3.3,3.4,])*(1e-3)
voltage_drop = np.array([1.1, 1.3, 1.5, 1.7, 2.0, 2.2, 2.4, 2.5, 2.6, 2.8, 3.1, 3.2, 3.3, 3.3, 3.3, 3.4, 3.4, 3.5, 3.5, 3.5])*(1e-3)
R_0 = 548000
```

```python
R_diode = R_0*bias_voltage/voltage_drop - R_0
R_diode
```

```python
plt.plot(bias_voltage, R_diode)
plt.title("Diode resistance vs bias voltage")
plt.xlabel('Bias voltage')
plt.ylabel('R diode')
plt.grid()
```

```python
plt.plot(-1*bias_voltage, -1*1000000000*voltage_drop/R_0)
plt.title("Leackage current vs bias voltage")
plt.xlabel('Bias voltage [V]')
plt.ylabel('Leackage current [nA]')
plt.grid()
plt.savefig("IV-scan-V_lt_0.png")
```

```python
plt.plot(voltage_drop/R_0*R_diode, 1000000000*voltage_drop/R_0)
plt.title("Voltage applied to diode [V] VS Current measured [nA]")
plt.grid()
plt.xlabel('Voltage applied [V]')
plt.ylabel('Current measured [nA]')
plt.savefig("figure1.png")
```
